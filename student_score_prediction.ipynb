# Load Dataset

```python
# fetch dataset 
student_performance = fetch_ucirepo(id=320) 
  
# data (as pandas dataframes) 
X = student_performance.data.features 
y = student_performance.data.targets 
  
df = pd.concat([X,y],axis=1)

df.head(5)
```

# Data Cleaning

- Check missing values
- Rename columns for clarity
- Convert categorical variables using Label Encoding or One-Hot Encoding

```python
def check_data(df):
    print("Data Check Summary: \n")
    
    # 1. Shape of the dataset
    print(f'Dataset contates {df.shape[0]} rows and {df.shape[1]} columns.\n')
    
    #Check missing values
    print('Missing Values per column:')
    missing = df.isnull().sum()
    
    if missing.any():
        print(missing[missing>0])
    else:
        print("No missing values found")
        
    # Duplicate rows
    duplicate_count = df.duplicated().sum()
    if duplicate_count > 0:
        print(f'found {duplicate_count}duplicate rows.\n')
    else:
        print('No duplicate rows found.\n')
```

```python
check_data(df)
```

The dataset contains no missing values, duplicates. This means that we don't have to resolve issues of missing values or duplicates. We will now move on to exploratory data analysis.
